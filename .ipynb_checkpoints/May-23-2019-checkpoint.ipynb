```python
import pandas as pd
DF=pd.read_csv("BULLMARKET.csv")
DF[:3]
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=np.array(DF["Date"])
Y=np.array(DF["Open"])
```

```python
plt.plot(X,Y)
plt.xlabel("Year")
plt.ylabel("Dow Jones Index")
plt.grid()
```

```python
plt.scatter(X,Y)
plt.xlim(2010,2010.125)
plt.ylim(9500,11000)
plt.gcf().set_size_inches(6,6)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X=X.reshape(-1,1)
xtrain, xtest, ytrain, ytest = train_test_split(X,Y)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
LR = LinearRegression() # instantiate linear regression
```

```python
LR # just to see what LR looks like, don't need this line
```

```python
model=LR.fit(xtrain,ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
YP=LR.predict(xtest)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(YP,ytest) # closer to 1 is better
```

```python
mean_squared_error(YP,ytest)
```

```python
np.sqrt(mean_squared_error(YP,ytest))
```

```python
YPall=LR.predict(X) # predicted value at each original point
```

```python
plt.plot(X,YPall-Y) # prediction error at each data point
plt.grid()
plt.plot([2009,2020],[1138.8, 1138.8],c="magenta")
plt.plot([2009,2020],[-1138.8, -1138.8],c="magenta")
plt.title("Residuals: Prediction Error")
```

```python
nruns=1000
RMSVALS=[]
R2VALS=[]
for j in range(nruns):
    xtrain, xtest, ytrain, ytest = train_test_split(X,Y)
    model=LR.fit(xtrain,ytrain)
    YP=LR.predict(xtest)
    RMS=np.sqrt(mean_squared_error(YP,ytest))
    R2=r2_score(YP,ytest) # closer to 1 is better
    RMSVALS.append(RMS)
    R2VALS.append(R2)
```

```python
info=plt.hist(RMSVALS,edgecolor="black",facecolor="pink",
             bins=np.linspace(1075,1225,12))
plt.xlabel("RMS")
```

```python
info
```

```python
plt.hist(R2VALS,edgecolor="black",facecolor="pink")
plt.xlabel("R2")
```

```python
plt.boxplot(R2VALS)
```

```python
nruns=100
RMSVALS=[]
R2VALS=[]
ALLRMS=[]
ALLR2=[]
X=X.ravel()
Y=Y.ravel()
for degree in range(2,15):
    RMSVALS=[]
    R2VALS=[]
    for j in range(nruns):
        xtrain, xtest, ytrain, ytest = train_test_split(X,Y)
        coefficients = np.polyfit(xtrain,ytrain,degree)
        YP=np.polyval(coefficients, xtest)
        RMS=np.sqrt(mean_squared_error(YP,ytest))
        R2=r2_score(YP,ytest) # closer to 1 is better
        RMSVALS.append(RMS)
        R2VALS.append(R2)
    ALLRMS.append(RMSVALS)
    ALLR2.append(R2VALS)
    
```

```python
plt.boxplot(ALLRMS);
```

```python
plt.boxplot(ALLR2);
```
