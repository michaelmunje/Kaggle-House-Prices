```python
import pandas as pd
import numpy as np
from mikes_utils import *
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
sale_prices = df['SalePrice']
df = df.drop(columns=['SalePrice'])
train_num = len(df)
df = pd.concat((df, df_test))
df.head()
```

```python
print_nan_cols(df, moderate_only=True)
```

```python
def add_none_cat(series):
    series = series.astype('category')
    series = series.cat.add_categories(['None'])
    series = series.fillna('None')
    return series
```

```python
df['Alley'] = add_none_cat(df['Alley'])
df['PoolQC'] = add_none_cat(df['PoolQC'])
df['MiscFeature'] = add_none_cat(df['MiscFeature'])
df['FireplaceQu'] = add_none_cat(df['FireplaceQu'])
df['Fence'] = add_none_cat(df['Fence'])
```

```python
print_nan_cols(df)
```

```python
df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
df['GarageType'] = add_none_cat(df['GarageType'])
df['GarageFinish'] = add_none_cat(df['GarageFinish'])
df['GarageQual'] = add_none_cat(df['GarageQual'])
df['GarageCond'] = add_none_cat(df['GarageCond'])
```

```python
df['ContainsGarage'] = df['GarageArea'].apply(lambda x: 1 if x == 0 else 0)
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
df['GarageArea'] = df['GarageArea'].fillna(0)
df['GarageCars'] = df['GarageCars'].fillna(0)
```

```python
df['BsmtFinSF1'] = df['BsmtFinSF1'].fillna(0)
df['BsmtFinSF2'] = df['BsmtFinSF2'].fillna(0)
df['BsmtUnfSF'] = df['BsmtUnfSF'].fillna(0)
df['TotalBsmtSF'] = df['TotalBsmtSF'].fillna(0)
df['BsmtFullBath'] = df['BsmtFullBath'].fillna(0)
df['BsmtHalfBath'] = df['BsmtHalfBath'].fillna(0)
```

```python
df['BsmtQual'] = add_none_cat(df['BsmtQual'])
df['ContainsBsmt'] = df['BsmtQual'].apply(lambda x: 1 if x == 'None' else 0)
df['BsmtCond'] = add_none_cat(df['BsmtCond'])
df['BsmtExposure'] = add_none_cat(df['BsmtExposure'])
df['BsmtFinType1'] = add_none_cat(df['BsmtFinType1'])
df['BsmtFinType2'] = add_none_cat(df['BsmtFinType2'])
```

```python
print_nan_cols(df)
```

```python
# df['MasVnrType'] = add_none_cat(df['MasVnrType'])
df['MasVnrType'] = df['MasVnrType'].fillna('None')
df['ContainsMasVnr'] = df['MasVnrType'].apply(lambda x: 1 if x == 'None' else 0)
df['MSSubClass'] = add_none_cat(df['MSSubClass'])
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
```

```python
df['MSZoning'] = df['MSZoning'].fillna(df['MSZoning'].mode()[0])
df['KitchenQual'] = df['KitchenQual'].fillna(df['KitchenQual'].mode()[0])
df['Exterior1st'] = df['Exterior1st'].fillna(df['Exterior1st'].mode()[0])
df['Exterior2nd'] = df['Exterior2nd'].fillna(df['Exterior2nd'].mode()[0])
df['SaleType'] = df['SaleType'].fillna(df['SaleType'].mode()[0])
```

```python
df = df.drop(columns=['Utilities'])
```

```python
df["Functional"] = add_none_cat(df['Functional'])
```

```python
df["Electrical"] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
df['MSSubClass'] = df['MSSubClass'].astype('category')
df['OverallCond'] = df['OverallCond'].astype('category')
df['MoSold'] = df['MoSold'].astype('category')
```

```python
to_encode = ['FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 
        'YrSold', 'MoSold']

for enc in to_encode:
    df[enc] = df[enc].astype('category')
```

```python
df['TotalFlrSF'] = df['1stFlrSF'] + df['2ndFlrSF']
```

```python
df['TotalSF'] = df['TotalBsmtSF'] + df['TotalFlrSF']
```

```python
df = adjust_skewness(df)
```

```python
df = convert_categorical_to_numbers(df)
```
